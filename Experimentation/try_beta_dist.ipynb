```python
import pyro
```

```python
#plot beta distributions with different parameters. Use a kdeplot for the density
import seaborn as sns


import matplotlib.pyplot as plt

def plot_beta(a, b):
    sns.kdeplot(pyro.distributions.Beta(a, b).sample([100_000]).numpy())
    plt.title(f'Beta({a}, {b})')
    plt.show()

def plot_beta_hist(a, b):
    sns.histplot(pyro.distributions.Beta(a, b).sample([100_000]).numpy(), bins=100)
    plt.title(f'Beta({a}, {b})')
    plt.show()
```

```python
plot_beta(0.1, 0.1)
```

```python
plot_beta_hist(1e-2,1e-2)
```
