```python
import torch 
import matplotlib.pyplot as plt

N = 100
P = 50
N_SAMPLES = 100_000

N_BATCHES = 1000
```

```python
def flatten_and_plot(tensor, bins = 1000):
    plt.hist(tensor.flatten().numpy(), bins=1000)
    plt.show()
```

```python
r1 = torch.randn(N_SAMPLES, P)

flatten_and_plot(r1)
```

```python
torch.manual_seed(0)
```

```python
r2 = []
N_SAMPLES_PER_BATCH = N_SAMPLES // N_BATCHES
for _ in range(N_BATCHES):
    r2.append(torch.randn(N_SAMPLES_PER_BATCH, P))

r2 = torch.cat(r2, dim=0)

flatten_and_plot(r2)
```
